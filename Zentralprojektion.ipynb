```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2
from mpl_toolkits.mplot3d.art3d import Line3DCollection
```

Hier soll die Funktion des Parallelepipeds dargestellt werden. Beim ersten Plot handelt es sich um die Darstellung des Parallelepipeds im Raum. Der zweite Plot stellt dann die Zentralprojektion des Parallelepipeds auf die xy Ebene dar.

Die Berechnung der Schnittpunkte mit der xy Ebene (zweite Plot) soll anhand des Punktes A aufgezeigt werden, für die weiteren Punkte des Parallelepipeds ist die Berechnung analog. ax, ay, az sind die Koordinaten des Punktes A und (x-az), (y-ay), (z-az) sind die Koordinaten des Richtungsvektors, wobei x, y und z die Koordinaten der Kamera sind. Rechts der Gleichung ist die xy Ebene dargestellt. Es handelt sich also um eine Gerade, die die xy Ebene schneidet.<br><br>
$\left( \begin{array}{r}
ax   \\                                              
ay  \\  
az   \\ 
\end{array}\right)$ + s 
$\left( \begin{array}{r}
x-ax   \\                                              
y-ay  \\  
z-az   \\ 
\end{array}\right)$ = 
$\left( \begin{array}{r}
x   \\                                              
y  \\  
0   \\ 
\end{array}\right)$
<br><br>
Es kann also die dritte Gleichung genommen werden und nach s aufgelöst werden.
<br><br>
s = $\dfrac{-az}{z-az}$ 
<br><br>
Dieser Wert kann anschliessend in die zwei oberen Gleichungen eingegeben werden, so erhält man die Koordinaten des Schnittpunkts des Punktes A mit der xy Ebene. Die genauen Berechnungen können weiter unten ab Zeile 113 studiert werden.

```python
####################################
# Funktion des Parallelepiped

# Hier wird das Parallelepiped gezeichnet
# Der erste Plot stellt das Parallelepiped im Raum dar
# Ein weiterer Plot die Zentralprojektion auf die xy-Ebene mit demselben Sehwinkel wie beim ersten Plot
# Der Sehwinkel kann bei Zeile 106 und 237 angepasst werden

def zentral(vax,vay,vaz,vbx,vby,vbz,vcx,vcy,vcz,x,y,z):
    plt.figure(figsize=(11,7))
    custom=plt.subplot(111,projection='3d')
    # Berechnung der Punkte des Parallelepiped
    # Der erste Punkt A des Parallelepipeds beginnt bei den Koordinaten (0.5,0.5,0.5)
    ax = 0.5
    ay = 0.5
    az = 0.5
    # Punkt B, Punkt A + Vektor a
    bx = float(ax) + float(vax)
    by = float(ay) + float(vay)
    bz = float(az) + float(vaz)
    # Punkt C, Punkt A + Vektor a + Vektor b
    cx = ax + vax + vbx
    cy = ay + vay + vby
    cz = az + vaz + vbz
    # Punkt D, Punkt A + Vektor b
    dx = ax + vbx
    dy = ay + vby
    dz = az + vbz
    # Punkt E, Punkt A + Vektor c
    ex = ax + vcx
    ey = ay + vcy
    ez = az + vcz
    # Punkt F, Punkt A + Vektor a + Vektor c
    fx = ax + vax + vcx
    fy = ay + vay + vcy
    fz = az + vaz + vcz
    # Punkt G, Punkt A + Vektor a + Vektor b + Vektor c
    gx = ax + vax + vbx + vcx
    gy = ay + vay + vby + vcy
    gz = az + vaz + vbz + vcz
    # Punkt H, Punkt A + Vektor b + Vektor c
    hx = ax + vbx + vcx
    hy = ay + vby + vcy
    hz = az + vbz + vcz
    
    # Zeichnen der Seiten des Parallelepipeds
    # Die Seite AB
    l1 = [(float(ax), float(ay), float(az)), (float(bx), float(by), float(bz))]
    line1 = Line3DCollection([l1], color='blue')
    # Die Seite BC
    l2 = [(float(bx), float(by), float(bz)), (float(cx), float(cy), float(cz))]
    line2 = Line3DCollection([l2], color='red')
    # Die Seite CD
    l3 = [(float(cx), float(cy), float(cz)), (float(dx), float(dy), float(dz))]
    line3 = Line3DCollection([l3], color='green')
    # Die Seite AD
    l4 = [(float(ax), float(ay), float(az)), (float(dx), float(dy), float(dz))]
    line4 = Line3DCollection([l4], color='gray')
    # Die Seite AE
    l5 = [(float(ax), float(ay), float(az)), (float(ex), float(ey), float(ez))]
    line5 = Line3DCollection([l5], color='magenta')
    # Die Seite BF
    l6 = [(float(bx), float(by), float(bz)), (float(fx), float(fy), float(fz))]
    line6 = Line3DCollection([l6], color='gold')
    # Die Seite CG
    l7 = [(float(cx), float(cy), float(cz)), (float(gx), float(gy), float(gz))]
    line7 = Line3DCollection([l7], color='lime')
    # Die Seite DH
    l8 = [(float(dx), float(dy), float(dz)), (float(hx), float(hy), float(hz))]
    line8 = Line3DCollection([l8], color='darkorange')
    # Die Seite EF
    l9 = [(float(ex), float(ey), float(ez)), (float(fx), float(fy), float(fz))]
    line9 = Line3DCollection([l9], color='slateblue')
    # Die Seite FG
    l10 = [(float(fx), float(fy), float(fz)), (float(gx), float(gy), float(gz))]
    line10 = Line3DCollection([l10], color='salmon')
    # Die Seite EH
    l11 = [(float(ex), float(ey), float(ez)), (float(hx), float(hy), float(hz))]
    line11 = Line3DCollection([l11], color='dodgerblue')
    # Die Seite HG
    l12 = [(float(hx), float(hy), float(hz)), (float(gx), float(gy), float(gz))]
    line12 = Line3DCollection([l12], color='lightcoral')
    # Hinzufügen der Seiten zum Plot
    plt.gca().add_collection3d(line1)
    plt.gca().add_collection3d(line2)
    plt.gca().add_collection3d(line3)
    plt.gca().add_collection3d(line4)
    plt.gca().add_collection3d(line5)
    plt.gca().add_collection3d(line6)
    plt.gca().add_collection3d(line7)
    plt.gca().add_collection3d(line8)
    plt.gca().add_collection3d(line9)
    plt.gca().add_collection3d(line10)
    plt.gca().add_collection3d(line11)
    plt.gca().add_collection3d(line12)
    # Bestimmung und Beschriftung der Achsen
    custom.set_title("Parallelepiped im Raum")
    custom.set_xlabel('X')
    custom.set_ylabel('Y')
    custom.set_zlabel('Z')
    # Sehwinkel
    # Für einen Blick senkrecht auf die xy Ebene, bitte azim=0 und elev=90 setzen
    custom.view_init(azim=60, elev=60)
    # Achsendimensionen bestimmen
    plt.xlim(-3, 3) 
    plt.ylim(-3, 3)
    plt.show()
    
    # Projektion der Punkte und Linien auf die xy Ebene
    plt2.figure(figsize=(10,7))
    custom2=plt2.subplot(111,projection='3d')
    
    # Schnittpunkt Richtungsvektor durch Punkt A mit xy Ebene
    # az + s * (z-az) = 0
    # s = (-az) / (z-az)
    sa= -az / (z-az)
    # x,y, und z Koordiante
    axp = ax + sa*(x-ax)
    ayp = ay + sa*(y-ay)
    azp = 0.0
    # Schnittpunkt Richtungsvektor durch Punkt B mit xy Ebene
    # bz + s * (z-bz) = 0
    # s = (-bz) / (z-bz)
    sb= -bz / (z-bz)
    # x,y, und z Koordiante
    bxp = bx + sb*(x-bx)
    byp = by + sb*(y-by)
    bzp = 0.0
    # Schnittpunkt Richtungsvektor durch Punkt C mit xy Ebene
    # cz + s * (z-cz) = 0
    # s = (-cz) / (z-cz)
    sc= -cz / (z-cz)
    # x,y, und z Koordiante
    cxp = cx + sc*(x-cx)
    cyp = cy + sc*(y-cy)
    czp = 0.0
    # Schnittpunkt Richtungsvektor durch Punkt D mit xy Ebene
    # dz + s * (z-dz) = 0
    # s = (-dz) / (z-dz)
    sd= -dz / (z-dz)
    # x,y, und z Koordiante
    dxp = dx + sd*(x-dx)
    dyp = dy + sd*(y-dy)
    dzp = 0.0
    # Schnittpunkt Richtungsvektor durch Punkt E mit xy Ebene
    # ez + s * (z-ez) = 0
    # s = (-ez) / (z-ez)
    se= -ez / (z-ez)
    # x,y, und z Koordiante
    exp = ex + se*(x-ex)
    eyp = ey + se*(y-ey)
    ezp = 0.0
    # Schnittpunkt Richtungsvektor durch Punkt F mit xy Ebene
    # fz + s * (z-fz) = 0
    # s = (-fz) / (z-fz)
    sf= -fz / (z-fz)
    # x,y, und z Koordiante
    fxp = fx + sf*(x-fx)
    fyp = fy + sf*(y-fy)
    fzp = 0.0
    # Schnittpunkt Richtungsvektor durch Punkt G mit xy Ebene
    # gz + s * (z-gz) = 0
    # s = (-gz) / (z-gz)
    sg= -gz / (z-gz)
    # x,y, und z Koordiante
    gxp = gx + sg*(x-gx)
    gyp = gy + sg*(y-gy)
    gzp = 0.0
    # Schnittpunkt Richtungsvektor durch Punkt H mit xy Ebene
    # hz + s * (z-hz) = 0
    # s = (-hz) / (z-hz)
    sh= -hz / (z-hz)
    # x,y, und z Koordiante
    hxp = hx + sh*(x-hx)
    hyp = hy + sh*(y-hy)
    hzp = 0.0
    
    # Zeichnen der Seiten des Parallelepipeds auf der xy Ebene
    # Alle z-Werte sind natürlich 0!
    # Die Seite AB
    l13 = [(float(axp), float(ayp), float(azp)), (float(bxp), float(byp), float(bzp))]
    line13 = Line3DCollection([l13], color='blue')
    # Die Seite BC
    l14 = [(float(bxp), float(byp), float(bzp)), (float(cxp), float(cyp), float(czp))]
    line14 = Line3DCollection([l14], color='red')
    # Die Seite CD
    l15 = [(float(cxp), float(cyp), float(czp)), (float(dxp), float(dyp), float(dzp))]
    line15 = Line3DCollection([l15], color='green')
    # Die Seite AD
    l16 = [(float(axp), float(ayp), float(azp)), (float(dxp), float(dyp), float(dzp))]
    line16 = Line3DCollection([l16], color='gray')
    # Die Seite AE
    l17 = [(float(axp), float(ayp), float(azp)), (float(exp), float(eyp), float(ezp))]
    line17 = Line3DCollection([l17], color='magenta')
    # Die Seite BF
    l18 = [(float(bxp), float(byp), float(bzp)), (float(fxp), float(fyp), float(fzp))]
    line18 = Line3DCollection([l18], color='gold')
    # Die Seite CG
    l19 = [(float(cxp), float(cyp), float(czp)), (float(gxp), float(gyp), float(gzp))]
    line19 = Line3DCollection([l19], color='lime')
    # Die Seite DH
    l20 = [(float(dxp), float(dyp), float(dzp)), (float(hxp), float(hyp), float(hzp))]
    line20 = Line3DCollection([l20], color='darkorange')
    # Die Seite EF
    l21 = [(float(exp), float(eyp), float(ezp)), (float(fxp), float(fyp), float(fzp))]
    line21 = Line3DCollection([l21], color='slateblue')
    # Die Seite FG
    l22 = [(float(fxp), float(fyp), float(fzp)), (float(gxp), float(gyp), float(gzp))]
    line22 = Line3DCollection([l22], color='salmon')
    # Die Seite EH
    l23 = [(float(exp), float(eyp), float(ezp)), (float(hxp), float(hyp), float(hzp))]
    line23 = Line3DCollection([l23], color='dodgerblue')
    # Die Seite HG
    l24 = [(float(hxp), float(hyp), float(hzp)), (float(gxp), float(gyp), float(gzp))]
    line24 = Line3DCollection([l24], color='lightcoral')
    # Hinzufügen der Seiten zum Plot
    plt2.gca().add_collection3d(line13)
    plt2.gca().add_collection3d(line14)
    plt2.gca().add_collection3d(line15)
    plt2.gca().add_collection3d(line16)
    plt2.gca().add_collection3d(line17)
    plt2.gca().add_collection3d(line18)
    plt2.gca().add_collection3d(line19)
    plt2.gca().add_collection3d(line20)
    plt2.gca().add_collection3d(line21)
    plt2.gca().add_collection3d(line22)
    plt2.gca().add_collection3d(line23)
    plt2.gca().add_collection3d(line24)
    # Bestimmung und Beschriftung der Achsen
    custom2.set_title("Projektion auf die xy Ebene mit Kamera")
    custom2.set_xlabel('X')
    custom2.set_ylabel('Y')
    custom2.set_zlabel('Z')
    # Sehwinkel auf die xy-Ebene
    # Für einen Blick senkrecht auf die xy Ebene, bitte azim=0 und elev=90 setzen
    custom2.view_init(azim=60, elev=60)
    # Achsendimensionen bestimmen
    plt2.xlim(-3, 3) 
    plt2.ylim(-3, 3)
    plt2.show()
```

Hier soll nun die Funktion aufgerufen werden. Mitgegeben werden die Koordinaten der Vektoren a, b und c. Am Ende werden auch noch die Koordinaten der Kamera mitgegeben.


```python
#########################################
# Das Parallelepiped wird als Aufbau mit Vektoren der Funktion übergeben
# Vektor a, Vektor b und Vektor c
# Die Kamera (x,y,z) wird ebenfalls mitgegeben, die letzten drei Werte entsprechen den Koordinaten.
# zentral(vax,vay,vaz,vbx,vby,vbz,vcx,vcy,vcz,x,y,z)
zentral(float(0.1),float(-0.7),float(0.4),float(0.4),float(0.3),float(-0.7),float(0.7),float(-0.4),float(0.3),float(2),float(2),float(2))
```
