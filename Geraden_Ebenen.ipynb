# Untersuchung, ob die Gerade das Dreieck schneidet
Hier wird nun der Schnittpunkt betrachtet.
<br>
<font color='red'>WICHTIG: Wie am Anfang erwähnt geht es hier nun um eine Gerade. Ob dann die Strecke denselben Schnittpunkt hat wird weiter unten betrachtet.</font>

Anhand eines Gleichungssytems mit drei Unbekannten, soll der Schnittpunkt ermittelt werden:

Das Gleichsetzen von Ebene und Gerade: A + s * (B-A) + t * (C-A) = P + r * (Q-P)

Das Umgestalten der Gleichung: s * (B-A) + t * (C-A) - r * (Q-P) = P - A

Die drei Gleichungen mit drei Unbekannten, wobei x, y und z den eingegebenen Koordinaten der eingegebenen Punkte entspricht:
<br>
1) s * (Bx - Ax) + t * (Cx - Ax) - r * (Qx - Px) = (Px - Ax)
<br>
2) s * (By - Ay) +  t * (Cy - Ay) - r * (Qy - Py) = (Py - Ay)
<br>
3) s * (Bz - Az) +  t * (Cz - Az) - r * (Qz - Pz) = (Pz - Az)

Dieses Gleichungssystem wird anhand von solve gelöst werden.

<font color='red'>1)</font> Um das Gleichungssystem lösen zu können, müssen die Werte vorbereitet werden.
<br><br>
Zuerst bestimmt werden die Werte (B - A), (C - A), (Q - P) und (P - A)
<br><br>
Anschliessend werden die Werte in a und b eingegeben und mit solve gelöst. Die Resultate werden in x als Array gespeichert.
<br><br>
x[0] entspricht dabei der Variable s, x[1] entspricht der Variable t und x[2] entspricht der Variable r.
<br><br>
Diese erhaltenen Werte werden zu Kontrolle einnmal in die Ebenengleichung und zweitens in die Geradengleichung gegeben.
<br><br>
Sollten diese Werte nun gleich sein, so gibt es einen Schnittpunkt. Ob die Gerade nun in der Dreiecksebene liegt muss überprüft werden. Zuest soll jedoch gesehen werden, ob sich der Schnittpunkt im Dreieck befindet oder nicht.
<br><br>
<font color='red'>2)</font> Damit der Schnittpunkt im Dreieck liegt, muss folgende Gleichung stimmen: s + t <=1
<br><br>
WARUM? Das Maximum, das erreicht werden kann ist wenn s=1 und t=0 ist oder s=0 und t=1 ist. Dann liegt der Schnittpunkt auf dem Punkt des Dreiecks. Alles unter und bis 1 sind Vektorkombinationen innerhalb des Dreiecks.
<br><br>
<font color='red'>3)</font> Sollte nun der Richtungsvektor der Geraden mit einem der Richtingsvektoren der Ebene übereinstimmen, liegt die Gerade in der Dreiecksebene und hat somit mehr als einen Schnittpunkt. Sollte dies nicht der Fall sein, so gibt es nur den einen Schnittpunkt, der mit Sxg, Syg und Szg angegeben wird.

```python
# Bestimmung, ob die Gerade das Dreieck schneidet oder nicht
# Schnittpunkt berechnen
# Anhand eines Gleichungssytems mit drei Unbekannten, soll der Schnittpunkt ermittelt werden
# Das Gleichsetzen von Ebene und Gerade: A + s * (B-A) + t * (C-A) = P + r * (Q-P)
# Das Umgestalten der Gleichung: s * (B-A) + t * (C-A) - r*(Q-P) = P - A
# Das Aufsetzen der drei Gleichungen
# Bestimmung der Zahlenwerte für die Gleichung
# (B-A)
gs1 = float(xb) - float(xa)
gs2 = float(yb) - float(ya)
gs3 = float(zb) - float(za)

# (C-A)
gt1 = float(xc) - float(xa)
gt2 = float(yc) - float(ya)
gt3 = float(zc) - float(za)

# (Q-P)
gr1 = -1*(float(xq) - float(xp))
gr2 = -1*(float(yq) - float(yp))
gr3 = -1*(float(zq) - float(zp))


# (P-A)
gax= float(xp) - float(xa)
gay= float(yp) - float(ya)
gaz= float(zp) - float(za)


# Werte in die Gleichung eingeben
# Ausgabe für s,t und r
a = np.array([[gs1,gt1,gr1], [gs2,gt2,gr2], [gs3,gt3,gr3]])
b = np.array([gax,gay,gaz])
x = solve1(a, b)


# Schnittpunkt mit der Ebene berechnen
# S = A + s * (B-A) + t * (C-A)
Sx = float(xa) + round(x[0],3)*gs1 + round(x[1],3)*gt1
Sy = float(ya) + round(x[0],3)*gs2 + round(x[1],3)*gt2
Sz = float(za) + round(x[0],3)*gs3 + round(x[1],3)*gt3 


# Schnittpunkt mir der Geraden berechnen
# S = P + r * (Q - P)
Sxg = float(xp) + round(x[2],3)*(-gr1)
Syg = float(yp) + round(x[2],3)*(-gr2)
Szg = float(zp) + round(x[2],3)*(-gr3)


# Sollte Sx = Sxg, Sy = Syg und  Sz = Szg sein, so handelt es sich um einen
# gemeinsamen Schnittpunkt
if round(Sx)==round(Sxg) and round(Sy)==round(Syg) and round(Sz)==round(Szg):
    print("Es gibt mindestens einen Schnittpunkt! Ob nun die Gerade auf der Dreiecksebene liegt, wird weiter überprüft!")
    print("Als erstes soll jedoch gesehen werden, ob der Schnittpunkt im Dreieck liegt.")
    print("")
    # Bestimmung, ob der Schnittpunkt im Dreieck liegt oder nicht
    # OS = OA + s(AB) + t(AC), wobei s und s € R+ und s+t ≤ 1. 
    # Da s und t bereits bestimmt sind, muss nur noch die Gleichung s+t ≤ 1 berücksichtigt werden
    i = round(x[0],3)+round(x[1],3)
    if i <=1:
        print("Der Schnittpunkt liegt im Dreieck!")
        print("")
        # Sollte der Richtungsvektor der  Geraden mit einem der Richtungsvektoren der Dreiecksebene
        # übereinstimmen, so liegt die Gerade in der Dreiecksebene
        if gs1==-gr1 and gs2==-gr2 and gs3==-gr3:
            print("Die Gerade liegt in der Dreiecksebene!")
        elif gt1==-gr1 and gt2==-gr2 and gt3==-gr3:
            print("Die Gerade liegt in der Dreiecksebene!")
        else:
            print("Es gibt nur den einen Schnittpunkt!")
            print("x-Wert:", Sxg)
            print("y-Wert:", Syg)
            print("z-Wert:", Szg)
            print("")
    else:
        print("Der Schnittpunkt liegt nicht im Dreieck!")
else:
    print("Es gibt keinen gemeinsamen Schnittpunkt!")
```

# Untersuchung, ob die Strecke das Dreieck schneidet
Hier wird nun der Schnittpunkt betrachtet.
<font color='red'>WICHTIG: Wie am Anfang erwähnt geht es hier nun um die Strecke.</font>
<br><br>
Mit der Hesse Normalform kann die Lage der beiden Punkte P und Q bestimmt werden. Sollten für beide Punkte <font color='red'>d(P,E)> 0</font> gelten, so sind beide Punkte oberhalb des Dreiecks und somit gibt es keinen Schnittpunkt. Ebenso be <font color='red'>d(P,E)< 0</font> sind beide Punkte unterhalb des Dreiecks und somit gibt es auch keinen Schnittpunkt. Sollte der eine Wert grösser/kleiner als der andere sein, so befinden sich die Punkte auf unterschiedlichen Seiten. ob nun die Strecke dann das Dreieck schneidet, muss weiter untersucht werden. Es handelt sich dabei um die zwei Normalenvektoren einer Ebene, die eine entgegengesetzte Lage besitzen.
<br><br>
Dies ist die Hesse Normalform:
<br><br>
E : $\dfrac{ax + bx + cx + d}{\sqrt{a^{2} + b^{2} + c^{2}}}$ = 0
<br><br>
Um jedoch die Hesse Normalform anwenden zu können, müssen wir die Parametergleichung des Dreiecks in die Koordinatengleichung überführen.
<br><br>
Die Umrechung sieht so aus:
$\left( \begin{array}{r}
u1 \\                                              
u2 \\
u3 \\ 
\end{array}\right)$ x 
$\left( \begin{array}{r}
v1  \\                                              
v2  \\
v3  \\ 
\end{array}\right)\circ$
$\left( \begin{array}{r}
x - px\\                                              
y - py \\
z - pz \\ 
\end{array}\right)$ = 0
<br><br>
Wobei u und v die Richtungsvektoren sind. px, py und pz entsprechen den Koordinaten des Punktes A.
<br><br>
    <font color='red'>1)</font> Das Kreuzprodukt der beiden Richtungsvektoren muss bestimmt werden:
$\left( \begin{array}{r}
u1 \\                                              
u2 \\
u3 \\ 
\end{array}\right)$ x 
$\left( \begin{array}{r}
v1  \\                                              
v2  \\
v3  \\ 
\end{array}\right)$ =
$\left( \begin{array}{rr}
u2*v3 - u3*v2  \\                                              
u3*v1 - u1*v3  \\  
u1*v2 - u2*v1  \\    
\end{array}\right)$
<br><br>
    <font color='red'>2)</font> Das Skalarprodukt bestimmen:
$\left( \begin{array}{rr}
u2*v3 - u3*v2  \\                                              
u3*v1 - u1*v3  \\  
u1*v2 - u2*v1  \\    
\end{array}\right)\circ$
$\left( \begin{array}{rr}
x - px  \\                                              
y - py  \\ 
z - pz  \\    
\end{array}\right)$ = 0
<br><br>
Ausgerechnet ist dies:
<br><br>
(u2v3 - u3v2)*x - px*(u2v3 - u3v2) + (u3v1 - u1v3)*y - py*(u3v1 - u1v3) + (u1v2 - u2v1)*z - pz*(u1v2 - u2v1) = 0
<br><br>
Umpositioniert ist dies:
<br><br>
(u2v3 - u3v2)*x + (u3v1 - u1v3)*y + (u1v2 - u2v1)*z  = px*(u2v3 - u3v2) + py*(u3v1 - u1v3) + pz*(u1v2 - u2v1)
<br><br>
a = (u2v3 - u3v2)
<br><br>
b = (u3v1 - u1v3)
<br><br>
c = (u1v2 - u2v1)
<br><br>
d =  px*(u2v3 - u3v2) + py*(u3v1 - u1v3) + pz*(u1v2 - u2v1)
<br><br><br>
Nun kann der Divisor und der Dividend der Hesse Normalform bestimmt werden. Als erstes soll überprüft werden, ob der Punkt P sich oberhalb oder unterhalb des Dreiecks befindet. Sollte die Rechnung dividend/divisor grösser sein als null, so befindet er sich oberhalb, ansonsten unterhalb. Auch mit dem Punkt Q wird so fortgefahren.
<br><br>
Nun folgt die Fallunterscheidung:
<br><br>
    <font color='red'>1)</font> Sollten sich beide oberhalb befinden, gibt es keinen Schnittpunkt. Der womöglich vorhin berechnete Schnittpunkt hängt damit zusammen, dass es sich um eine Gerade gehandelt hat. Für die Strecke gilt dies jedoch nicht.
<br><br>
    <font color='red'>2)</font> Sollten sich beide unterhalb befinden, gibt es keinen Schnittpunkt. Der womöglich vorhin berechnete Schnittpunkt hängt damit zusammen, dass es sich um eine Gerade gehandelt hat. Für die Strecke gilt dies jedoch nicht.
<br><br>
    <font color='red'>3)</font> Sollte sich P oberhalb und Q unterhalb befinden und es gilt s+t<=1, so befindet sich der Schnittpunkt auf dem Dreieck. Q wird mit gestrichelter blauer Farbe bis zum Schnittpunkt gezeichnet, da sich dieser Teil unterhalb befindet und P wird mit ganzer Linie bis zum Schnittpunkt gezeichnet, da sich dieser Teil oberhalb befindet.
<br><br>
    <font color='red'>4)</font> Sollte sich Q oberhalb und P unterhalb befinden und es gilt s+t<=1, so befindet sich der Schnittpunkt auf dem Dreieck. P wird mit gestrichelter blauer Farbe bis zum Schnittpunkt gezeichnet, da sich dieser Teil unterhalb befindet und Q wird mit ganzer Linie bis zum Schnittpunkt gezeichnet, da sich dieser Teil oberhalb befindet.

```python
# Bestimmung, ob die Strecke das Dreieck schneidet
# Mit der Hesse Normalform kann bestimmt werden, ob sich die Punkte beide oberhalb, unterhalb oder oberhalb und unterhalb befinden.
# Sollte sich ein Teil unterhalb befinden, wird dieser gestrichelt angezeigt.

# Zuerst soll anhand den gegebenen Punkten A, B und C die Koordinatengleichung bestimmt werden
# Sprich die Umrechung der Parameterdarstellung in die Koordinatendarstellung
# Parameter a,b,c und d
koordx = gs2*gt3 - gs3*gt2
koordy = gs3*gt1 - gs1*gt3
koordz = gs1*gt2 - gs2*gt1
# d muss hier für die Hesse Normalform überprüft werden.
# Da d auch auf der gleichen Seite der anderen Parameter geschrieben werden muss,
# müssen je nach dem das Vorzeichen gewechselt werden.
d = float(xa)*(koordx) + float(ya)*(koordy) + float(za)*(koordz)
if d<0:
    d = d*-1
if d>0:
    d = d*-1


# Hier wird der untere Teil des Bruches vorbereitet
divisor = math.sqrt(koordx**2 + koordy**2 + koordz**2)

# Hier wird überprüft ob P sich unterhalb oder oberhalb des Dreiecks befindet
# Bei d(P,E) > 0 liegt auf der Seite der Ebene in der der Normalenvektor zeigt
# Bei d(P,E) < 0 liegt auf der anderen Seite
dividend = koordx*float(xp) + koordy*float(yp) + koordz*float(zp) + d


if dividend/divisor > 0:
    po=1
    pu=0
    print("P befindet sich oberhalb!")
else:
    po=0
    pu=1
    print("P befindet sich unterhalb!")

# Hier wird überprüft ob Q sich unterhalb oder oberhalb des Dreiecks befindet
# Bei d(Q,E) > 0 liegt auf der Seite der Ebene in der der Normalenvektor zeigt
# Bei d(Q,E) < 0 liegt auf der anderen Seite
dividend = koordx*float(xq) + koordy*float(yq) + koordz*float(zq) + d


if dividend/divisor > 0:
    qo=1
    qu=0
    print("Q befindet sich oberhalb!")
else:
    qo=0
    qu=1
    print("Q befindet sich unterhalb!")

# Hier wird nun ausgegeben, wo sich die beiden Punkte befinden
if po==1 and qo==1:
    print("Beide Punkte sind oberhalb und deswegen gibt es auch keinen Schnittpunkt!")
    print("Wohlgemerkt: Womöglich hat das Programm bereits einen Schnittpunkt ausgegeben.")
    print("Dies jedoch nur, weil es sich um eine Gerade durch die Punkte handelt.")
    print("Hier ist es eine Strecke und somit gibt es keinen Schnittpunkt.")
elif pu==1 and qu==1:
    print("Beide Punkte sind unterhalb und deswegen gibt es auch keinen Schnittpunkt!")
    print("Wohlgemerkt: Womöglich hat das Programm bereits einen Schnittpunkt ausgegeben.")
    print("Dies jedoch nur, weil es sich um eine Gerade durch die Punkte handelt.")
    print("Hier ist es eine Strecke und somit gibt es keinen Schnittpunkt.")
elif po==1 and qu==1:
    print("P ist oberhalb, Q unterhalb!");
    print("Die Punkte befinden sich auf verschiedenen Seiten!")
    print("Bitte schauen Sie für einen potenziellen Schnittpunkt die Berechnungen mit der Geraden an!")
    print("Siehe dazu die Berechnungen von oben!")
    print("Denn die Ergebnisse der Gerade und Strecke sind hier gleichwertig!")
    ####################################
    # Hier sollen nun das Dreieck und die Strecke erneut gezeichnet werden
    # Gemäss den Anforderungen soll der untere Teil gestrichelt werden
    # Q bis Schnittpunkt gestrichelt
    # Schnittpunkt bis P gerade
    if i <=1:
        # Grösse des Plots
        plt3.figure(figsize=(10,7))
        custom3=plt3.subplot(111,projection='3d')
        # l2 ist unterhalb des Schnittpunkts bis zum Schnittpunkt
        l2 = [(float(xq), float(yq), float(zq)), (float(Sxg), float(Syg), float(Szg))]
        # l3 ist vom Schnittpunkt bis oberhalb des Schnittpunkt
        l3 = [(float(Sxg), float(Syg), float(Szg)), (float(xp), float(yp), float(zp))]
        # Erneut das Dreieck
        x12=np.array([float(xa), float(xb), float(xc)])
        y12=np.array([float(ya), float(yb), float(yc)])
        z12=np.array([float(za), float(zb), float(zc)])
        custom3.scatter(x12,y12,z12)
        verts2 = [list(zip(x12, y12, z12))]
        # Formatierung von Dreieck und Linien
        triangle3 = Poly3DCollection(verts2, alpha=.45, facecolor='green')
        line2 = Line3DCollection([l2], color='blue', linestyle='--')
        line3 = Line3DCollection([l3], color='red')
        # Hinzufügen des Objekte in den Plot
        plt3.gca().add_collection3d(triangle3)
        plt3.gca().add_collection3d(line2)
        plt3.gca().add_collection3d(line3)
        # Bestimmen der Achsen
        custom3.set_title("Lineare Algebra 1a")
        custom3.set_xlabel('X')
        custom3.set_ylabel('Y')
        custom3.set_zlabel('Z')
        # Sehwinkel 90 Grad auf die xy-Ebene
        custom3.view_init(azim=0, elev=90)
        # Achsenlimitationen
        plt3.xlim(-1, 7) 
        plt3.ylim(-1, 7)
        plt3.show()
    else:
        print("Der Schnittpunkt liegt nicht im Dreieck!")
elif pu==1 and qo==1:
    print("P ist unterhalb, Q ist oberhalb!")
    print("Die Punkte befinden sich auf verschiedenen Seiten!")
    print("Bitte schauen Sie für einen potenziellen Schnittpunkt die Berechnungen mit der Geraden an!")
    print("Siehe dazu die Berechnungen von oben!")
    print("Denn die Ergebnisse der Gerade und Strecke sind hier gleichwertig!")
    ####################################
    # Hier sollen nun das Dreieck und die Strecke erneut gezeichnet werden
    # Gemäss den Anforderungen soll der untere Teil gestrichelt werden
    # P bis Schnittpunkt gestrichelt
    # Schnittpunkt bis Q gerade
    if i <=1:
        # Grösse des Plots
        plt3.figure(figsize=(10,7))
        custom3=plt3.subplot(111,projection='3d')
        # l2 ist unterhalb des Schnittpunkts bis zum Schnittpunkt
        l2 = [(float(xp), float(yp), float(zp)), (float(Sxg), float(Syg), float(Szg))]
        # l3 ist vom Schnittpunkt bis oberhalb des Schnittpunkt
        l3 = [(float(Sxg), float(Syg), float(Szg)), (float(xq), float(yq), float(zq))]
        # Erneut das Dreieck
        x12=np.array([float(xa), float(xb), float(xc)])
        y12=np.array([float(ya), float(yb), float(yc)])
        z12=np.array([float(za), float(zb), float(zc)])
        custom3.scatter(x12,y12,z12)
        verts2 = [list(zip(x12, y12, z12))]
        # Formatierung von Dreieck und Linien
        triangle3 = Poly3DCollection(verts2, alpha=.45, facecolor='green')
        line2 = Line3DCollection([l2], color='blue', linestyle='--')
        line3 = Line3DCollection([l3], color='red')
        # Hinzufügen des Objekte in den Plot
        plt3.gca().add_collection3d(triangle3)
        plt3.gca().add_collection3d(line2)
        plt3.gca().add_collection3d(line3)
         # Bestimmen der Achsen
        custom3.set_title("Lineare Algebra 1a")
        custom3.set_xlabel('X')
        custom3.set_ylabel('Y')
        custom3.set_zlabel('Z')
        # Sehwinkel 90 Grad auf die xy-Ebene
        custom3.view_init(azim=0, elev=90)
        # Achsenlimitationen
        plt3.xlim(-1, 7) 
        plt3.ylim(-1, 7)
    else:
        print("Der Schnittpunkt liegt nicht im Dreieck!")
else:
    print("Fehler")
```
